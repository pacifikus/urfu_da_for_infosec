### FE

```python
df_hourly = data.set_index('timestamp').resample('H').mean().reset_index()
df_daily = data.set_index('timestamp').resample('D').mean().reset_index()
```

```python
for df in [df_hourly, df_daily]:
    df['Weekday'] = (pd.Categorical(df['timestamp'].dt.strftime('%A'), categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday', 'Saturday', 'Sunday']))
    df['Hour'] = df['timestamp'].dt.hour
    df['Day'] = df['timestamp'].dt.weekday
    df['Month'] = df['timestamp'].dt.month
    df['Year'] = df['timestamp'].dt.year
    df['Month_day'] = df['timestamp'].dt.day
    df['Lag'] = df['value'].shift(1)
    df['Rolling_Mean'] = df['value'].rolling(7, min_periods=1).mean()
    df = df.dropna()
```

```python
df_hourly = (df_hourly
             .join(df_hourly.groupby(['Hour','Weekday'])['value'].mean(),
                   on = ['Hour', 'Weekday'], rsuffix='_Average')
            )

df_daily = (df_daily
            .join(df_daily.groupby(['Hour','Weekday'])['value'].mean(),
                  on = ['Hour', 'Weekday'], rsuffix='_Average')
           )

df_hourly.tail()
```

```python
df_hourly = df_hourly.dropna()
df_daily_model_data = df_daily[['value', 'Hour', 'Day',  'Month','Month_day','Rolling_Mean']].dropna()

model_data = df_hourly[['value', 'Hour', 'Day', 'Month_day', 'Month','Rolling_Mean','Lag', 'timestamp']].set_index('timestamp').dropna()
model_data.head()
```

### Isolation Forest

```python
model = IsolationForest(
    random_state=2024,
    contamination=0.005,
    n_estimators=200,
    max_samples=0.7
)

model.fit(model_data)
```

```python
df_hourly['Outliers'] = pd.Series(model.predict(model_data)).apply(lambda x: 1 if x == -1 else 0)
score = model.decision_function(model_data)
df_hourly['Score'] = score
df_hourly.head()
```

### KMeans

```python
df = df_hourly.copy()
```

```python
cols = ['value', 'Hour', 'Day', 'Month', 'Year', 'Month_day', 'Lag', 'Rolling_Mean', 'value_Average']

n_cluster = range(2, 11)

kmeans = [KMeans(n_clusters = i).fit(df[cols]) for i in n_cluster]
scores = [kmeans[i].inertia_ for i in range(len(kmeans))]
```

### Статистические методы

image.png

*   25th percentile (Q1)
*   75th percentile (Q3)
*   interquartile range (IQR = Q3 – Q1)
*   Lower inner fence: Q1 – (1.5 * IQR)
*   Upper inner fence: Q3 + (1.5 * IQR)
*   Lower outer fence: Q1 – (3 * IQR)
*   Upper outer fence: Q3 + (3 * IQR)

```python
def stats_method(df, variable):
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1
    inner_fence = 1.5 * iqr
    outer_fence = 3 * iqr

    inner_fence_le = q1 - inner_fence
    inner_fence_ue = q3 + inner_fence

    outer_fence_le = q1 - outer_fence
    outer_fence_ue = q3 + outer_fence

    outliers = []
    for index, x in enumerate(df[variable]):
        if x <= inner_fence_le or x >= inner_fence_ue:
            outliers.append(1)
        else:
            outliers.append(0)
    return outliers
```

```python
outliers = stats_method(df, "value")
```

```python
df['outlier'] = outliers
```

```python
hover = HoverTool(tooltips=tooltips)

hv.Points(
    df.query("outlier == 1")).opts(
        size=10,
        color='#ff0000'
    ) * hv.Curve(df).opts(
        opts.Curve(
            title="New York City Taxi Demand Anomalies",
            xlabel="",
            ylabel="Demand",
            height=300,
            responsive=True,tools=[hover, 'box_select', 'lasso_select', 'tap'],
            show_grid=True,
        )
    )
```
